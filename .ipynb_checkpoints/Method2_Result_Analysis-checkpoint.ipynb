```python
import scipy.stats as stats 
import json
```

## Opening the file

```python
with open("logProbScore_BEST.json", "r") as f:
    data = json.load(f)
```

```python
data
```

## Checking Accuracy

```python
def compare(a, b):
     return 1 if a < b else 0

def calAccu(option):
    # Counters for total and correct comparisons
    total_comparisons = 0
    correct_comparisons = 0
    # Result dictionary to store comparison results
    result = {}
    
    # Iterate through the data
    for key, values in data.items():
        comparisons = {}

        if option == 1 or option == "total":
        # Perform comparisons if the keys exist
            if 'least' in values and 'more' in values:
                res = compare(values['least'], values['more'])
                comparisons['least_vs_more'] = res
                total_comparisons += 1
                correct_comparisons += res

        if option == 2 or option == "total": 
            if 'more' in values and 'most' in values:
                res = compare(values['more'], values['most'])
                comparisons['more_vs_most'] = res
                total_comparisons += 1
                correct_comparisons += res
            
        if option == 3 or option == "total":
            if 'least' in values and 'most' in values:
                res = compare(values['least'], values['most'])
                comparisons['least_vs_most'] = res
                total_comparisons += 1
                correct_comparisons += res
    
        result[key] = comparisons
    
    # Calculate percentage of correct comparisons
    accuracy_percentage = (correct_comparisons / total_comparisons) * 100 if total_comparisons > 0 else 0
    # Display the results
    print("Comparison Results:", result)
    print(f"Accuracy: {accuracy_percentage:.2f}% ({correct_comparisons}/{total_comparisons} correct)")

    return result

calAccu(3)
```

## Running Statistical Significance Test - Wilcoxon Sign Ranked Test

```python
# Collect "least" and "most" scores
least_scores = []
most_scores = []

for key, values in data.items():
    if 'least' in values and 'most' in values:
        least_scores.append(values['least'])
        most_scores.append(values['most'])

# Ensure we have enough data points
if len(least_scores) > 1:
    # Perform Wilcoxon Signed-Rank Test
    stat, p_value = stats.wilcoxon(least_scores, most_scores, alternative='less')

    # Print results
    print(f"Wilcoxon test statistic: {stat}")
    print(f"P-value: {p_value}")

    # Interpret result
    alpha = 0.05  # Significance level (5%)
    if p_value < alpha:
        print("Reject H₀: 'Least' scores are statistically significantly smaller than 'Most' scores.")
    else:
        print("Fail to reject H₀: No significant difference between 'Least' and 'Most' scores.")
else:
    print("Not enough data for a statistical test.")
```
